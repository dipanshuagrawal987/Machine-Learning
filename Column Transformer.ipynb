```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('covid_toy.csv')
df.head()
```

```python
df[df['fever'].isnull()]
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
si.fit(df)
```

```python
df = pd.DataFrame(si.transform(df),columns=df.columns)
```

```python
df['cough'].value_counts()
```

```python
cough = pd.DataFrame(df['cough'])
cough.sample(5)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Mild','Strong']])
```

```python
oe.fit(cough)
```

```python
cough = oe.transform(cough)

# Encoded array of the column cough
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False,dtype=int,drop='first')
```

```python
ohe.fit(df[['gender']])
```

```python
gender = ohe.transform(df[['gender']])

# encoded array of column gender
gender
```

```python
ohe.fit(df[['city']])
```

```python
city = ohe.transform(df[['city']])

# encoded array of column city
```

```python
df
```

```python
age = np.array(df[['age']])
fever = np.array(df[['fever']])
has_covid = np.array(df[['has_covid']])
```

```python
new_array = np.hstack([age, gender, fever, cough, city, has_covid])
```

# The above hectic code can be done in the few lines using **Column Transformer**

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(missing_values=np.nan, strategy='most_frequent'),['fever']),
    ('tnf2',OneHotEncoder(sparse_output=False,dtype=int,drop='first'),['city','gender']),
    ('tnf3',OrdinalEncoder(categories=[['Mild','Strong']]),['cough'])
] ,remainder='passthrough')
```

```python
transformer.fit_transform(df)
```

# One Hot Encoding using pandas

```python
pd.get_dummies(df,columns=['fuel','owner'], dtype = int)
```

## K-1 Encoding (dropping the first column from the resulting columns formed)

```python
pd.get_dummies(df,columns=['fuel','owner'], dtype = int, drop_first=True)
```

# One Hot Encoding using Sklearn (Recommended to see the documentation of anything where doubt may arise)

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1],test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first', dtype = int)
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
x_train_new

# by default a sparse matrix is formed thus we need to convert it to numpy array
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']]).toarray()
x_train_new

# To convert sparse matrix to numpy array we use .toarray() function
# We can also ignore this if we train OneHotEncoding() using sparse=False parameter
```

```python
x_train_rest = np.array(x_train.iloc[:,0:2])
x_train_rest
```

```python
x_train = np.hstack((x_train_rest, x_train_new))
x_train
```
